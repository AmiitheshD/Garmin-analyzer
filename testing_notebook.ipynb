```python
import pandas as pd

data_frame = pd.read_csv("garmin_data.csv")
```

```python
# Show DataFrame info (don't wrap in print — info() prints and returns None)
data_frame.info()
```

```python
# Normalize column names and drop only columns that exist
data_frame.columns = data_frame.columns.str.strip()
cols_to_drop = ["Favorite","Calories","Avg Stride Length","Avg GAP","Normalized Power® (NP®)","Training Stress Score®","Title","Aerobic TE","Activity Type","Number of Laps", "Decompression", "Steps","Elapsed Time", "Moving Time","Best Lap Time"]
missing = [c for c in cols_to_drop if c not in data_frame.columns]
print('Missing columns:', missing)
to_drop = [c for c in cols_to_drop if c in data_frame.columns]
data_frame.drop(columns=to_drop, inplace=True)
```
